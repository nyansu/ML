```python
from sklearn.preprocessing import StandardScaler
import numpy as np

df = dataset
X_train = df[['squareMeters', 'basement', 'attic', 'garage']]
X_train
```

```python
# норм. числовых признаков с помощью разных методов 
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
print('Мат ожидание', scaler.mean_)
print('Дисперсия', scaler.var_)
print('Преобразованный набор')
arr = scaler.transform(X_train)
DfTr = pd.DataFrame(arr, columns=['squareMeters', 'basement', 'attic', 'garage'])
DfTr
```

```python
df = dataset
#df['category'] = dataset['category'].astype('category')
df['price_type'] = dataset['price'].apply(lambda price: 'Низкий' if price < 2_500_000  else 'Средний' if price < 5_000_000 else 'Высокий').astype('category')
df
```

```python

df.dtypes
```

```python
from sklearn.compose import make_column_selector as selector

categorical_columns_selector = selector(dtype_include='category')
categorical_columns = categorical_columns_selector(df)
categorical_columns
```

```python
df_cat = df[categorical_columns] #отберем только те столбцы в которых присутствует категориальная перемнная
df_cat.head(20)
```

```python
df["price_type"].value_counts().sort_index()
```

```python
# преобразуем данные в числовое представление
from sklearn.preprocessing import OrdinalEncoder

cat_column = df_cat[["price_type"]]

print(df_cat['price_type'].unique()) #посмотрим какие виды образования присутствуют в выборке

encoder = OrdinalEncoder()   # вызываем кодировщик
encoder  = encoder.fit(df_cat[['price_type']])
df_cat.loc[:,'price_type'] = encoder.transform(df_cat[['price_type']])
df_cat
```

```python
encoder = OrdinalEncoder()
price_type_encoded = encoder.fit_transform(df[['price_type']])

print(f"Encoded price_type column:\n{price_type_encoded}")

encoded_categories = encoder.categories_[0]

for i, category in enumerate(encoded_categories):
    print(f"Значение {i} принадлежит к категории: {category}")
```
